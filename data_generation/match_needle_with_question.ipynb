```python
import pandas as pd
```

```python
prev_qa = pd.read_excel("Fresh_QA_1.xlsx", sheet_name="Really Filtered")
prev_qa
```

```python
new_qa = pd.read_excel("FreshQA_v03312025.xlsx")
new_qa
```

```python
new_qa.columns = new_qa.iloc[1]
new_qa = new_qa[2:].reset_index(drop=True)
new_qa
```

```python
for _, new_row in new_qa.iterrows():
    match = prev_qa[prev_qa['question'] == new_row['question']]
    if not match.empty:
        prev_qa.loc[match.index, 'needle'] = new_row['answer_0']
```
